# 9.정렬

- 순서 없이 배열된 자료들을 값에 따라 순서에 따라 재배열하는 것
- 키 : 자료를 정렬하는데 사용하는 자료
- 정렬의 효율성 : 얼마만큼 빠르게 정렬을 실시하는지
- 정렬의 안정성 : 같은 키 값을 가지는 자료들을 입력 순서 그대로 정렬하는 지
- 다중키 정럴   : 정렬에 사용되는 키가 여러 개 있을 때

## 1. 정렬의 종류

- 정렬이 실행되는 장소에 따라
    1. 내부 정렬 : 컴퓨터 메모리 내부에서 정렬
    2. 외부 정렬 : 외부 보조 기억 장치에서 정렬
    
    
- 정렬 알고리즘에 따라
    1. 교환 : 키 값을 비교, 자료를 교환 -> 버블, 퀵 정렬
    2. 삽입 : 키 값을 비교, 자료를 삽입 -> 삽입, 셀 정렬
    3. 병합 : 키 값을 비교, 자료를 병합 -> 2-way, n-way 병합
    4. 분배 : 키 값을 여러 개의 부분 집합으로 분배, 이를 이용하여 정렬 -> 기수 정렬
    5. 선택 : 트리 자료구조 등 특정 자료구조를 통하여 정렬 -> 히프 정렬

## 2. 선택 정렬

- 정렬되지 않은 전체 자료 중에서 해당 위치에 맞는 자료를 선택하여 위치를 교환


- 방법(자료 개수 n, 정렬 위치 k)
    1. k 번째로 작은 수를 (n-k+1)개에서 찾아서 k번째로 교환시킨다.
    2. 1을 마지막까지 반복한다.
    
    
- 연산 시간
    - 루프문 실행    : O(n^2)
    - 자료 교환      : O(n)
    - 최선,평균,최악 : O(n^2 + n) = O(n^2)
        
- 특징       
    - 자료의 이동 연산 횟수가 적기 때문에 자료 크기가 큰 정렬에 유리
    - 효율성이 낮다.
    - 안정성을 만족하지 않는다.

## 3. 버블 정렬

- 정렬되지 않은 자료들을 대상으로 인접한 두 개 자료의 키 값을 비교하여 위치를 교환하는 정렬


- 방법
    - 처음부터 다음 위치의 키 값과 비교하여, 다음 위치 키 값이 작다면 위치를 교환
    - 교환이 이루어지지 않을때까지 반복
    
    
- 연산 시간
    - 루프문 실행 : O(n^2)
    - 자료 이동
        - 최선         : O(0) = 0
        - 평균 및 최악 : O(n^2) <자료가 여가순으로 정렬 되어있을 경우>
    - 최선,평균,최악 : O(n^2)
    
    
- 특징
    - 느린 알고리즘
    - 자료가 클 때는 사용하기 어렵다.
    - 안정성은 유지가 가능

## 4. 퀵 정렬

- 중심 값(피봇)을 기준으로 두 자료의 키 값을 비교하여 위치를 교환하는 정렬
- 분할 정복 기법이다.

- 방법
    - Left, Rigth 변수가 자료 맨 처음과, 마지막을 가리킨다, 피봇은 가장 마지막을 가리킨다.
    - Left는 다음 위치에 자신보다 큰 수가 있다면 이동
    - Right는 이전 위치에 자신보다 작은 수가 있다면 이동한다.
    - Left, Right가 이동하지 못하게 되면 둘을 서로 바꾼뒤 위의 행동을 반복한다.
    - Left, Right가 같은 위치에 있다면, 그 위치와 피봇을 바꾼다.
    - 정렬된 피봇을 기준으로 분할 정복을 실시한다.
    
    
- 연산 시간
    - 최선, 평균 : O(nlog_{2}n)
    - 최악 : O(n^2)

## 5. 삽입 정렬
- 기존에 정렬된 부분 집합에 정렬할 자료의 위치를 찾아 삽입하는 정렬


- 연산 시간
    - 최선, 평균 : O(n)
    - 최악 : O(n^2)

## 6. 셸 정렬

- 기존 자료를 일정한 간격에 의해 여러 개의 부분 집합으로 나눈 다음, 각 부분 집합에 대해 삽입 정렬을 수행하는 정렬 방법
- 간격을 일반적으로 초기에는 전체 개수의 1/2로 하고 계속해서 1/2로 줄여나간 후 1이 될때까지 반복한다.

- 연산 시간
    - 최선 : O(n)
    - 평균 : O(n^1.25)
    - 최악 : O(n^20

## 7. 병합 정렬

- 같은 개수의 원소를 가지는 부분 집합으로 기존 자료를 분할하고 분할된 각 부분 집합을 병합하는 정렬

- 정렬된 자료 집합개수에 따라
    1. 2-way 병합 : 2개의 정렬된 자료 집합을 병합하는 경우
    2. n-way 병합 : n개의 정렬된 자료를 결합하는 경우
